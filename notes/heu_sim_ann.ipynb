```python
def generate_weighted_adjacency_matrix(nodes, total_connections):
    if total_connections > nodes * (nodes - 1) // 2:
        raise ValueError("Total connections exceed the maximum possible connections in a complete graph.")

    graph = np.zeros((nodes, nodes), dtype=object)

    # Create a list of all possible edge pairs
    all_possible_edges = [(i, j) for i in range(nodes) for j in range(i+1, nodes)]
    
    # Randomly choose edge pairs to form connections
    chosen_edges = random.sample(all_possible_edges, total_connections)
    
    for i, j in chosen_edges:
        x = random.randint(1, 8)
        y = random.randint(x + 1, 9)
        weight = [x, y]
        graph[i, j] = weight
        graph[j, i] = weight
    
    return graph
```

```python
# przedstawienie graficzne macierzy incydencji
def display_weighted_adjacency_matrix(matrix):
    for row in matrix:
        display_row = []
        for val in row:
            if val:
                display_row.append(f"[{val[0]}, {val[1]}]")
            else:
                display_row.append("0")
        print(display_row)
```

```python
# funkcja generująca wszystkie możliwe scenariusze pomiędzy wierzchołkami dla macierzy incydencji
def generate_weight_scenarios(adjacency_matrix):
    num_vertices = len(adjacency_matrix)
    weight_scenarios = []

    for _ in range(100):
        scenario = np.zeros_like(adjacency_matrix, dtype=int)

        for i in range(num_vertices):
            for j in range(num_vertices):
                if isinstance(adjacency_matrix[i][j], list) and len(adjacency_matrix[i][j]) == 2:
                    weight_range = adjacency_matrix[i][j]
                    weight = random.randint(weight_range[0], weight_range[1])
                    scenario[i][j] = weight

        weight_scenarios.append(scenario)

    return weight_scenarios
```

```python
def find_random_path(graph):
    def dfs(node, path):
        path = path + [node]
        if len(path) == len(graph):
            return [path]
        paths = []
        for neighbor, weight in enumerate(graph[node]):
            if weight != 0 and neighbor not in path:
                new_paths = dfs(neighbor, path)
                for new_path in new_paths:
                    paths.append(new_path)
        return paths

    all_paths = []
    for node in range(len(graph)):
        paths = dfs(node, [])
        all_paths.extend(paths)

    # Map node indices to alphabet letters
    letter_map = {i: chr(65 + i) for i in range(len(graph))}
    random_path = [letter_map[node] for node in random.choice(all_paths)]
    return random_path
```

```python
def get_random_path_with_weights(num_vertices, adjacency_matrix):
    vertices = string.ascii_uppercase[:num_vertices]
    path_length = random.randint(4, num_vertices-2)
    path = random.sample(vertices, path_length)
    weights = []
    for i in range(len(path) - 1):
        start_vertex = ord(path[i]) - ord('A')
        end_vertex = ord(path[i + 1]) - ord('A')
        weight = adjacency_matrix[start_vertex][end_vertex]
        if weight == 0:
            get_random_path_with_weights(num_vertices,adjacency_matrix)
        else:
            weights.append(weight)
    return path
```

```python
# funkcja obliczająca współczynnik żalu
def get_regret_ratio(cost, max_cost):
    regret_ratio = cost/max_cost
    return regret_ratio
```

```python
# funkcja obliczająca współczynnik żalu dla scenariuszy
def min_max_regret_for_scenarios(path,scenarios,temperature):
    min_cost = 999
    min_regret = 999
    max_cost = 0
    min_regret_ratio = 999
    
    for i in range(len(scenarios)):
        path_weights = get_path_weights(path,scenarios[i])
        cost = sum(path_weights)
        #print(cost)
        if min_cost > cost:
            min_cost = cost
            #print("zmniejszam min cost")
            #print(min_cost)
        if max_cost < cost :
            max_cost = cost
            #print("zwiekszam max cost")
            #print(max_cost)
        regret_ratio_for_scenario = get_regret_ratio(cost, max_cost)
        delta  = regret_ratio_for_scenario - min_regret_ratio
        #print(regret_ratio_for_scenario)
        if  delta < 0 :
            min_regret_ratio = regret_ratio_for_scenario
        else:
            if math.exp(delta / temperature) > random.random():
                min_regret_ratio = regret_ratio_for_scenario
        #print("zmniejszam min_regret_ratio")
        #print(min_regret_ratio)
    return min_regret_ratio
```

```python
# funkcja zwracająca scieżkę z najmniejszym współczynnikiem żalu
def get_min_ratio_path(data):
    paths = data['path']
    ratios = data['best_ratio']

    min_ratio_index = ratios.index(min(ratios))
    min_ratio = ratios[min_ratio_index]
    min_ratio_path = paths[min_ratio_index]

    return min_ratio, min_ratio_path
```

```python
# proces symulowanego wyrzażania
def simulated_annealing(number_of_nodes, total_connections, number_of_iterations, initial_temperature, cooling_coefficient):
    results = {'path': [], 'best_ratio': []}
    graph = generate_weighted_adjacency_matrix(number_of_nodes, total_connections)
    graph_nodes = get_alphabet_letters(number_of_nodes)
    display_weighted_adjacency_matrix(graph)
    scenarios = generate_weight_scenarios(graph)
    test_scenario = scenarios[1]
    path = find_random_path(test_scenario)
    print(path)
    try:
        for iteration in range(number_of_iterations):
            temperature = initial_temperature / (1 + iteration * cooling_coefficient)
            best_ratio = min_max_regret_for_scenarios(path, scenarios, temperature)
            results['path'].append(path)
            results['best_ratio'].append(best_ratio)
            path = change_vertex(path, test_scenario, results)
    except NoSuitableVerticesException:
        return get_min_ratio_path(results)

    return get_min_ratio_path(results)
```
